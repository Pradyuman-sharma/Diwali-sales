```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns 
import numpy as np
```

```python
ax = sns.countplot(x= 'Gender', data= df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen = df.groupby(['Gender'],as_index = False)['Amount'].sum().sort_values(by='Amount' ,ascending = False)
sales_gen
```

```python
sales_gen = df.groupby(['Gender'],as_index = False)['Amount'].sum().sort_values(by='Amount' ,ascending = False)
ax =sns.barplot(x='Gender', y='Amount', data = sales_gen)
```

```python
ax =sns.countplot(x='Age Group', data=df, hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age = df.groupby(['Age Group'],as_index = False)['Amount'].sum().sort_values(by='Amount' ,ascending = False)
sales_age
```

```python
sales_age = df.groupby(['Age Group'],as_index = False)['Amount'].sum().sort_values(by='Amount' ,ascending = False)
sns.barplot(x='Age Group', y = 'Amount', data = sales_age)
```

```python
satate = df.groupby(['State'],as_index = False)['Orders'].sum().sort_values(by='Orders' ,ascending = False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x='State', y = 'Orders', data = satate)
```

```python
satate = df.groupby(['State'],as_index = False)['Amount'].sum().sort_values(by='Amount' ,ascending = False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x='State', y = 'Amount', data = satate)
```

```python
grouped_df = df.groupby(['State', 'Age Group', 'Gender']).size().reset_index(name='order_count')

most_orders_df = grouped_df.loc[grouped_df.groupby(['State', 'Gender'])['order_count'].idxmax()].head(22)
sns.set(style="whitegrid")
ax = sns.barplot(x='State', y='order_count', hue='Gender', data=most_orders_df)
for bars in ax.containers:
    ax.bar_label(bars)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

```python
grouped_df = df.groupby(['State', 'Age Group', 'Gender'])['Amount'].sum().reset_index().head(100)


sns.set(style="whitegrid")

sns.barplot(x='State', y='Amount', hue='Gender', data=grouped_df,  errorbar=None)
plt.xticks(rotation=45)

plt.title('Comparison of Amount by Age Group, Gender, and State')
plt.xlabel('State')
plt.ylabel('Total Amount')
plt.legend(title='Gender')
g=sns.catplot(x='State', y='Amount', hue='Gender', col='Age Group', data=grouped_df, kind='bar',  errorbar=None, height=6, aspect=1.2)

plt.tight_layout()
plt.show()
```

```python
ax = sns.countplot(data = df, x = "Marital_Status")
sns.set(rc={'figure.figsize':(8,5)})
for bars in ax.containers:
    ax.bar_label(bars)
        
```

```python
sales_state = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data = sales_state, x = "Marital_Status", y ="Amount", hue="Gender")
```
